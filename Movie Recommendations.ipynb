```python
import numpy as np              #numpy short form-->np
import pandas as pd             ## same
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer #--> covert text to numreric value
from sklearn.metrics.pairwise import cosine_similarity #--> count the similarity score
```

```python
movies_data = pd.read_csv('/content/drive/My Drive/BCT Gen AI/Movies_data.csv') #-->convert csv file to pandas frame
movies_data.head()
```

# Before the final project

---



```python
movies_data.shape #number of rows&col
```

```python
#feature extraction
selected_features = ['genres', 'keywords', 'original_language', 'tagline', 'cast', 'director']
selected_features
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')  # replacing the null value to null string
```

```python
combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['original_language'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director']
```

```python
combined_features
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features) #print combine feature to the vector/numerical values
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors) #measure the similarity score
print(similarity)
```

```python
similarity.shape #1-->movie index,2-->similarity score
```

```python
#Taking user input for their fav movie name
movie_name = input('Enter your favourite movie name : ')
```

```python
list_of_all_titles = movies_data['title'].tolist() #print the title name
print(list_of_all_titles)
```

```python
#Getting close match of the movie given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
if len(find_close_match) == 0:
  print('No close match found')
else:
  print(find_close_match)
```

```python
close_match = find_close_match[0]  #print the fst match
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]  #print index of the movie
index_of_the_movie
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie])) #enumerate-->is a kind of loop
print(similarity_score)   #fst is index,similarity score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)    #sorting the similar movies in decending order
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  homepage_link = movies_data[movies_data.index==index]['homepage'].values[0]
  if (i<16):
    if pd.isna(homepage_link):
      print(i, '.',title_from_index, ' - page not available')
    else:
      print(i, '.',title_from_index, ' - ', homepage_link)
    i+=1
```

# Movie Recommendation System

---



```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

if len(find_close_match) == 0:
  print('No such movie or similar movies found!')

else:

  close_match = find_close_match[0]

  index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

  similarity_score = list(enumerate(similarity[index_of_the_movie]))

  sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

  print('\nMovies suggested for you : \n')

  i = 1

  for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    homepage_link = movies_data[movies_data.index==index]['homepage'].values[0]
    if (i<16):
      if pd.isna(homepage_link):
        print(i, '.',title_from_index, ' - page not available')
      else:
        print(i, '.',title_from_index, ' - ', homepage_link)
      i+=1
```
